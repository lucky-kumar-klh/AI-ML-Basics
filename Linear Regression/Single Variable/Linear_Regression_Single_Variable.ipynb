```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("area-price.csv")
```

```python
%%time
# %matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area, df.price, color = 'red', marker = '*')
```

# Linear Regression

```python
# Create an object for Linear Regression
reg = linear_model.LinearRegression()

# training LR model using fit()
reg.fit(df[['area']], df.price)

# y = mx + b --> Linear Regression Model
print("m (slope) = ", reg.coef_)
print("b (intercept) = ", reg.intercept_)

# y = (135.78767123 * x) + 180616.43835616432
```

```python
# price prediction
reg.predict([[3300], [5000]])   # parameter: 2D Array
```

```python
d = pd.read_csv("area.csv")

# applying dataframe d of other file to get predicted prices using our trained model reg
predicted_prices = reg.predict(d)  # Returns an array
print(predicted_prices)

d['Predicted Price'] = predicted_prices  # Adding a column in dataframe d
d.to_csv("Predicted Prices wrt Area.csv", index = False)  # Creating CSV with predicted prices
```

# Excercise

```python
# Creating a model object
income_prediction_model = linear_model.LinearRegression()

dataframe = pd.read_csv("canada_per_capita_income.csv")

# Training Model using fit(X, y): X - Dataframe, y - Series
income_prediction_model.fit(dataframe[['year']], dataframe.per_capita_income)

plt.xlabel("Year")
plt.ylabel("Canada Per Capita Income")
plt.scatter(dataframe.year, dataframe.per_capita_income, color = 'red')
# plt.plot(dataframe)

# Making a column for our predicted per capita income
predicted_values = income_prediction_model.predict(dataframe[['year']])

dataframe['Predicted Per Capita Income'] = predicted_values
dataframe.to_csv("Final_Task_Canada_per_Capita_Income.csv", index = False)

print("y = mx + b, where m =", income_prediction_model.coef_[0], "& b =", income_prediction_model.intercept_)
```

```python
print("Per Capita Income @2020 is", end = ' ')
income_prediction_model.predict([[2020]])[0]
```
